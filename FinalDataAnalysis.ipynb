# DESCRIBING THE DATA #

`Time Period` is the year in which activities were recorded

`Geography` Locations in New York City

`Number of Walking or Biking` raw count of people walking or biking in area

`Percent of Walking or Biking` percent count walking/biking

`Number Monthly Bicycle Use`raw count of monthly bicycle use in area

`Percent Monthly Bicycle Us`percent count bicycle use

`Number Recent Exercise` raw count of recent exercise

`Percent Recent Exercise` percent count recent exercise

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Combined_NYC_EH_Dataset.csv"  # Replace with your dataset path
df = pd.read_csv(file_path)

# Define the list of big cities or urban regions
urban_regions = [
    'Downtown - Heights - Slope',
    'Bedford Stuyvesant - Crown Heights',
    'Williamsburg - Bushwick',
    'Washington Heights',
    'Upper West Side',
    'East Harlem',
    'Long Island City - Astoria',
    'Central Harlem - Morningside Heights'
]

# Filter the dataset for these regions
urban_data = df[df['Geography'].isin(urban_regions)].copy()

# Convert 'Number_WalkingOrBiking' and 'Number_MonthlyBicycleUse' to numeric (removing commas)
urban_data['Number_WalkingOrBiking'] = pd.to_numeric(urban_data['Number_WalkingOrBiking'].str.replace(',', ''), errors='coerce')
urban_data['Number_MonthlyBicycleUse'] = pd.to_numeric(urban_data['Number_MonthlyBicycleUse'].str.replace(',', ''), errors='coerce')

# Aggregate data by region for the latest available year
latest_year = urban_data['TimePeriod'].max()
urban_latest_data = urban_data[urban_data['TimePeriod'] == latest_year]

# Plot the data: Number of Walking/Biking and Monthly Bicycle Use by Geography
plt.figure(figsize=(12, 6))

# Bar width settings
bar_width = 0.35
index = range(len(urban_latest_data['Geography']))

plt.bar(index, urban_latest_data['Number_WalkingOrBiking'], bar_width, label='Walking/Biking')
plt.bar(
    [i + bar_width for i in index],
    urban_latest_data['Number_MonthlyBicycleUse'],
    bar_width,
    label='Monthly Bicycle Use'
)

# Plot formatting
plt.xlabel('Urban Regions')
plt.ylabel('Number of People')
plt.title(f'Walking/Biking and Monthly Bicycle Use in Urban Regions ({latest_year})')
plt.xticks([i + bar_width / 2 for i in index], urban_latest_data['Geography'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Combined_NYC_EH_Dataset.csv"  # Replace with your dataset path
df = pd.read_csv(file_path)

# Define the list of big cities or urban regions
urban_regions = [
    'Downtown - Heights - Slope',
    'Bedford Stuyvesant - Crown Heights',
    'Williamsburg - Bushwick',
    'Washington Heights',
    'Upper West Side',
    'East Harlem',
    'Long Island City - Astoria',
    'Central Harlem - Morningside Heights'
]

# Filter the dataset for these regions
urban_data = df[df['Geography'].isin(urban_regions)].copy()

# Convert 'Number_WalkingOrBiking' and 'Number_MonthlyBicycleUse' to numeric (removing commas)
urban_data['Number_WalkingOrBiking'] = pd.to_numeric(urban_data['Number_WalkingOrBiking'].str.replace(',', ''), errors='coerce')
urban_data['Number_MonthlyBicycleUse'] = pd.to_numeric(urban_data['Number_MonthlyBicycleUse'].str.replace(',', ''), errors='coerce')

# Aggregate data by region for the latest available year
latest_year = urban_data['TimePeriod'].max()
urban_latest_data = urban_data[urban_data['TimePeriod'] == latest_year]

# Plot the data using box plots
plt.figure(figsize=(14, 6))

# Box plot for Walking/Biking
plt.subplot(1, 2, 1)
plt.boxplot(urban_latest_data['Number_WalkingOrBiking'].dropna(), vert=True, patch_artist=True)
plt.title(f'Box Plot for Number of Walking/Biking ({latest_year})')
plt.ylabel('Number of People')

# Box plot for Monthly Bicycle Use
plt.subplot(1, 2, 2)
plt.boxplot(urban_latest_data['Number_MonthlyBicycleUse'].dropna(), vert=True, patch_artist=True)
plt.title(f'Box Plot for Monthly Bicycle Use ({latest_year})')
plt.ylabel('Number of People')

# Display the plots
plt.tight_layout()
plt.show()
```
