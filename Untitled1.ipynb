OUR PLOT WILL HAVE:
- ROOMS as the X axis
- VALUE as the y axis
- WE WILL USE DISTANCE TO DISTINGUISH BETWEEN POINTS (e.g. Distance <5)

```python
X = Boston ['Rooms']
y = Boston ['Value']
distance = Boston ['Distance']
```

```python
above_threshold = distance > 3.0
below_threshold = distance <= 3.0
```

```python
plt.scatter(X, y, color='blue', label="house")

plt.xlabel("Rooms")
plt.ylabel("Value")
plt.title("Relationships between rooms and value")

plt.legend()
plt.show()
```
